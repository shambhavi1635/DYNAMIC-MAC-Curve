```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

pd.set_option("display.max_columns",None)
```

```python
df = pd.read_excel("notebooks/MarketPrices2010-2025.xlsx")

df.head()
```

```python
print("Shape :",df.shape)

print("\nColumns")

print(df.columns.tolist())

print("\nInformation")

df.info()
```

```python
df = df.dropna(axis=1,how="all")

print(df.shape)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))

sns.heatmap(df.isnull(),
            cbar=False,
            cmap="viridis")

plt.title("Missing Values")

plt.show()
```

```python
print("Duplicate Rows :",df.duplicated().sum())
```

```python
df.describe().T
```

```python
df.dtypes
```

```python
numerical_columns = df.columns[1:]

for column in numerical_columns:

    plt.figure(figsize=(7,4))

    plt.hist(df[column],bins=8)

    plt.title(column)

    plt.xlabel(column)

    plt.ylabel("Frequency")

    plt.show()
```

```python
for column in numerical_columns:

    plt.figure(figsize=(8,2))

    plt.boxplot(df[column],vert=False)

    plt.title(column)

    plt.show()
```

```python
for column in numerical_columns:

    plt.figure(figsize=(10,5))

    plt.plot(df["Year"],
             df[column],
             marker="o",
             linewidth=2)

    plt.title(column)

    plt.xlabel("Year")

    plt.grid(True)

    plt.show()
```

```python
plt.figure(figsize=(14,8))

for column in numerical_columns:

    plt.plot(df["Year"],
             df[column],
             marker="o",
             label=column)

plt.legend(bbox_to_anchor=(1.05,1))

plt.xlabel("Year")

plt.title("Market Price Trends")

plt.grid(True)

plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,10))

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Matrix")

plt.show()
```

```python
growth = df.copy()

for column in numerical_columns:

    growth[column] = growth[column].pct_change()*100

growth.head()
```

```python
growth.describe().T
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(df.set_index("Year"),
            cmap="YlGnBu",
            linewidth=0.5)

plt.title("Market Prices Across Years")

plt.show()
```

```python
for column in numerical_columns:

    plt.figure(figsize=(6,4))

    sns.kdeplot(df[column],
                fill=True)

    plt.title(column)

    plt.show()
```
